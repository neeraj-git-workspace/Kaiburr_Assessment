# **Text Pre-Processing**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#transforms the values in those columns by replacing each value with its corresponding integer label

encoder=LabelEncoder()
columns_to_encode = ['Sub-product',	'Issue',	'Sub-issue',	'Company',	'State',	'Consumer consent provided?',	'Submitted via',	'Company response to consumer',	'Timely response?',	'Category']
data[columns_to_encode] = data[columns_to_encode].apply(encoder.fit_transform)
```

```python
data.head()
```

```python
data.tail()
```

# **Multi Classification model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.metrics.pairwise import kernel_metrics
classifier1=DecisionTreeClassifier()
classifier2=MultinomialNB()
classifier3=KNeighborsClassifier()
classifier4=SVC(kernel='linear')

X=data.iloc[:,:-1]
y=data.iloc[:,-1]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#Decision Tree

classifier1.fit(X_train,y_train)
```

```python
#Naive Bayes

classifier2.fit(X_train,y_train)
```

```python
#K Nearest Neighbour

classifier3.fit(X_train,y_train)
```

```python
y_pred1=classifier1.predict(X_test)
y_pred2=classifier2.predict(X_test)
y_pred3=classifier3.predict(X_test)
```

# **Model Evaluation**

```python
print(classification_report(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
```

```python
print(classification_report(y_test, y_pred2))
print(confusion_matrix(y_test, y_pred2))
```

```python
print(classification_report(y_test, y_pred3))
print(confusion_matrix(y_test, y_pred3))
```

```python
models=['DecisionTree', 'NavieBayes', 'KNearestNeighbors']
accuracy1=accuracy_score(y_test, y_pred1)
accuracy2=accuracy_score(y_test, y_pred2)
accuracy3=accuracy_score(y_test, y_pred3)
accuracy=[accuracy1,accuracy2,accuracy3]
comparison=pd.DataFrame(accuracy, index=models)
comparison
```

```python
comparison.plot(kind='bar',legend=False)
```
